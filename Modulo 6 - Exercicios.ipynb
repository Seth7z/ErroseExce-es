```python
try:
  nome = input('Digite seu nome completo: ')
  idade = int(input('Digite sua idade: '))
  print(f"Nome informado foi: {nome}")
  print(f"Idade informada foi: {idade}")
except:
  print('Algo está errado')

else:
  print('tente novamente!')

finally:
    print("Programa finalizado!")
```
